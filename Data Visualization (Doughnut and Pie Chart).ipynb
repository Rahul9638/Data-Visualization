```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
fruits = ["Orange","Gauva","Apple","Pomegranate"]
quant = [3,32,55,33]
```

```python
plt.pie(quant,labels=fruits)
plt.show()
```

```python
# Changing asthetics of the pie chart i.e formating the looks of the pie chart
```

```python
plt.pie(quant,labels=fruits,autopct="%0.1f%%",colors=["orange","green","red","pink"])
plt.show()
```

```python
#Creating dough-nut chart....
#Basically it has a hole in the centre of the circle
```

```python
plt.pie(quant,labels=fruits,radius=2)
plt.pie([1],colors="w",radius=1)
plt.show()
```
