```python
# Adding multiple scatter plot
```

```python
a = [10,20,30,40,50,60,70,80]
b = [1,2,3,4,5,6,7,8]
c = [2,4,6,8,1,2,4,6]
plt.scatter(a,b,marker="*",c="y",s=100)
plt.scatter(a,c,marker="*",c="g",s=50)
plt.show()
```
